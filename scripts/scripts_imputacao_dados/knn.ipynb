## Preenchendo Dados com o Método de KNN do SKilearn
Os datasets utilizados foram os de vazão da data 10/06/2023 da rotas pa-ba e pr-am. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
from sklearn.impute import KNNImputer
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('../../datasets/vazao/10-06-2023/tratado/bbr/tratado bbr esmond data pa-ba 06-10-2023.csv')
df = df.drop(["Data", "Intervalo"], axis=1)

df["Vazao"] = df["Vazao"].replace(-1.0, np.nan)

scaler = MinMaxScaler()

df[["Vazao"]] = scaler.fit_transform(df[["Vazao"]])

imputer = KNNImputer(n_neighbors=245)

df = pd.DataFrame(imputer.fit_transform(df),columns = df.columns)

df[["Vazao"]] = scaler.inverse_transform(df[["Vazao"]])

df.head(15)
```

### Preenchendo os dados faltantes

```python
def preencher_knn(df):
    df_copy = df.copy()
    
    # Substitua os valores -1.0 por NaN
    df_copy['Vazao'] = df_copy['Vazao'].replace(-1.0, np.nan)
    
    # Transforme a coluna 'Vazao' em um array 2D, pois o KNNImputer espera isso.
    vazao_array = df_copy['Vazao'].values.reshape(-1, 1)
    
    imputer = KNNImputer(n_neighbors=5)
    
    # Ajuste e transforme os valores NaN usando o KNNImputer
    df_copy['Vazao'] = imputer.fit_transform(vazao_array)
    
    return df_copy
```
